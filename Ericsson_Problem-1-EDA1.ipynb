### 2. Sentiment analysis

```python
text_df = train_df[['summary', 'positives', 'negatives', 'overall']]
```

```python
text_df = text_df.replace(np.nan, 'review')
```

#### importing punctuations and stop words and parser

```python
import string
punctuations = string.punctuation
```

```python
from nltk.corpus import stopwords
english_stop_words = stopwords.words('english')
english_stop_words.remove('no')
english_stop_words.remove('not')
english_stop_words.append('review')
english_stop_words.remove('down')

english_stop_words.remove('too')
```

```python
import spacy
parser = spacy.load('en_core_web_md')
parser.Defaults.stop_words -= {'no', 'not', 'down', 'to', 'much'}
parser.Defaults.stop_words |= set(english_stop_words)
parser.Defaults.stop_words |= set(punctuations)

for word in parser.Defaults.stop_words:
    lex = parser.vocab[word]
    lex.is_stop = True
```

#### cleaning summary column

```python
def clean_summary(raw_data):
    parsed_data = parser(raw_data)
    temp_df = pd.DataFrame([(t.text, t.pos_, t.dep_, t.is_stop) for t in parsed_data], index=range(len(parsed_data)), columns=['text', 'POS', 'DEP', 'IS_STOP'])
    
    temp_df = temp_df[temp_df['IS_STOP']==False]
    temp_df = temp_df[temp_df['POS'].isin(['VERB','X','NUM']) == False]
    temp_df = temp_df.drop(temp_df[temp_df['POS']=='PROPN'][temp_df['DEP'].isin(['compound'])==False].index)
    
    ngram = []
    bag_of_words = []
    last_index = -1
    for idx, item in temp_df['text'].items():
        if idx-last_index == 1:
            ngram.append(item)
        else:
            if len(ngram):
                bag_of_words.append(' '.join(ngram))
                ngram.clear()
            ngram.append(item)
        last_index = idx
    bag_of_words.append(' '.join(ngram))
        
    return ','.join(bag_of_words)
```

```python
for i in range(10):
    print(text_df['summary'][i])
    print(clean_summary(text_df['summary'][i].lower()))
    print()
```

```python
new_df = text_df['summary'].apply(clean_summary)
```
